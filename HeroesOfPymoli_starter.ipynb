## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Clear duplicate information
age_demo_df = pd.DataFrame(purchase_data[['SN', 'Age','Purchase ID', 'Price']])
age_demo_df = age_demo_df.drop_duplicates()
age_demo_df
```

```python
#Making Bins for age ranges
Bins_age = [0, 9, 14, 19, 24, 29, 34, 39, 100]
Bin_Labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
#Categorize the players into the bins
age_demo_df['Age Group'] = pd.cut(age_demo_df['Age'], Bins_age, labels = Bin_Labels)
age_demo_df
```

```python
#Calculate how many players by age group and what percent each group has.
age_demo_df = age_demo_df.sort_values('Age Group')
age_demo = age_demo_df['Age Group'].value_counts()
percent_age = age_demo_df['Age Group'].value_counts() / sum(age_demo)
```

```python
#Summary table for demographics
summary_df = pd.DataFrame({'Player Count':age_demo,
                          'Total Percentage':percent_age})
summary_df['Total Percentage'] = pd.Series(['{0:.2f}%'.format(val * 100) for val in summary_df['Total Percentage']], index =
                                           summary_df.index)
summary_df = summary_df.sort_index(axis = 0)
summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Separate data for purchase count, average purchase price, and total purchase value by players
top_spender_df = pd.DataFrame(purchase_data[['SN', 'Price', 'Item Name']])
#Find top players
VIP_group = top_spender_df.groupby('SN')
#Get the total number of purchases by player
money_spent = VIP_group['Item Name'].count()
#Calculate the total purchace price
total_spent_pp = VIP_group['Price'].sum()
#find the average purchase price per player
avg_amount_spent = VIP_group['Price'].mean()
#Create the summary table from the data obtained
Spenders_summary = pd.DataFrame({'Purchase Count':money_spent,
                                'Average Purchase Price':avg_amount_spent,
                                'Total Purchase Value':total_spent_pp,})
#Sort the total purchase value column in descending order
Spenders_summary = Spenders_summary.sort_values(['Total Purchase Value'], ascending=False).head()
Spenders_summary
```
